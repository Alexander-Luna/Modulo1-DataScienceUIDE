## 6锔 Conexi贸n y consulta b谩sica en SQLite

Hemos usado la librer铆a `sqlite3` para crear una base de datos en memoria, definir una tabla, insertar datos y realizar una consulta b谩sica.

```python
import sqlite3

# Conectar a una base de datos en memoria (no se guarda al cerrar)
conn = sqlite3.connect(':memory:')
print("Conexi贸n a base de datos SQLite en memoria establecida.")

# Crear una tabla simple
cursor = conn.cursor()
cursor.execute("CREATE TABLE IF NOT EXISTS empleados (id INTEGER PRIMARY KEY, nombre TEXT, puesto TEXT)")
print("Tabla 'empleados' creada.")

# Insertar algunos datos
empleados_data = [
    ('Alice', 'Data Scientist'),
    ('Bob', 'Data Engineer'),
    ('Charlie', 'Analista BI')
]
cursor.executemany("INSERT INTO empleados (nombre, puesto) VALUES (?, ?)", empleados_data)
conn.commit() # Guardar los cambios
print("Datos insertados en la tabla 'empleados'.")

# Realizar una consulta
print("\nConsulta de todos los empleados:")
cursor.execute("SELECT nombre, puesto FROM empleados")
# Fetchall recupera todos los resultados de la consulta
resultados_sqlite = cursor.fetchall()
print(resultados_sqlite)

# Cerrar la conexi贸n
conn.close()
print("\nConexi贸n a SQLite cerrada.")

```

 **Explicaci贸n l铆nea por l铆nea:**

   ** conn = sqlite3.connect(':memory:')**
    
>   - 驴Qu茅 hace? Crea una base de datos SQLite temporal en memoria RAM (no se guarda en disco).

>   - 驴Por qu茅 usar :memory:? Es 煤til para pruebas, demostraciones o procesamiento r谩pido sin necesidad de un archivo f铆sico.

>    - Devuelve un objeto Connection llamado conn, que se usar谩 para interactuar con la base de datos.

   ** df_csv.to_sql('tips', conn, index=False)**
        
>   -    驴Qu茅 hace?  Toma el DataFrame df_csv (por ejemplo, tips.csv) y lo convierte en una tabla llamada tips dentro de la base de datos creada.
>   -    conn: es la conexi贸n que creamos previamente.
>   -    'tips': es el nombre de la tabla SQL que se crear谩.
>   -    index=False: evita que se agregue la columna de 铆ndice de pandas a la tabla SQL.

   ** consulta = pd.read_sql_query('SELECT sex, AVG(total_bill) as avg_bill FROM tips GROUP BY sex', conn)**

>   -     驴Qu茅 hace? Ejecuta una consulta SQL sobre la base SQLite y devuelve el resultado como un nuevo DataFrame de pandas.
>   -     SELECT sex, AVG(total_bill) as avg_bill: selecciona la columna sex y calcula el promedio de la columna total_bill, renombr谩ndola como avg_bill.
>   -     FROM tips: indica que la consulta se hace sobre la tabla tips.
>   -     GROUP BY sex: agrupa los registros por g茅nero (sex), por lo que obtendr谩s una fila por cada g茅nero (Male, Female, etc.) con su respectivo promedio de cuenta.
